## üìù Ïã§Ïäµ Î¨∏Ï†ú 1: Ï§ëÏã¨ Í≤ΩÌñ• Ï∏°ÎèÑ

### Î∞∞Í≤Ω
Ï∞®ÎüâÏùò **EMISSIONS(CO2 Î∞∞Ï∂úÎüâ)** Îç∞Ïù¥ÌÑ∞Î°ú Îã§ÏñëÌïú ÎåÄÌëúÍ∞íÏùÑ Í≥ÑÏÇ∞ÌïòÍ≥† ÎπÑÍµêÌï©ÎãàÎã§.

### Ïù¥Î°† Î≥µÏäµ

| ÎåÄÌëúÍ∞í | ÌäπÏßï | Ïù¥ÏÉÅÏπò ÎØºÍ∞êÎèÑ |
|--------|------|:-------------:|
| ÏÇ∞Ïà†ÌèâÍ∑† | Î™®Îì† Í∞í Î∞òÏòÅ | Îß§Ïö∞ ÎØºÍ∞ê |
| Ï§ëÏïôÍ∞í | Ï†ïÎ†¨ ÌõÑ Í∞ÄÏö¥Îç∞ Í∞í | Í∞ïÍ±¥Ìï® |
| ÏµúÎπàÍ∞í | Í∞ÄÏû• ÎπàÎ≤àÌïú Í∞í | - |
| Ï†àÏÇ¨ÌèâÍ∑† | ÏñëÎÅù Ï†úÍ±∞ ÌõÑ ÌèâÍ∑† | Ï§ëÍ∞Ñ |
| Í∞ÄÏ§ëÌèâÍ∑† | Í∞ÄÏ§ëÏπò Î∞òÏòÅ ÌèâÍ∑† | Îß§Ïö∞ ÎØºÍ∞ê |

### ÏàòÌñâÌï† ÏûëÏóÖ
1. EMISSIONSÏùò ÏÇ∞Ïà†ÌèâÍ∑†, Ï§ëÏïôÍ∞í, ÏµúÎπàÍ∞í Í≥ÑÏÇ∞
2. 10%, 20% Ï†àÏÇ¨ÌèâÍ∑† Í≥ÑÏÇ∞ Î∞è ÏÇ∞Ïà†ÌèâÍ∑†Í≥º ÎπÑÍµê
3. Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Ïù∏ Ï∞®ÎüâÏóê ÎåÄÌïú Í∞ÄÏ§ëÌèâÍ∑† Í≥ÑÏÇ∞ (COMPACT 50%, MID-SIZE 30%, SUV 20%)
4. Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Î≥Ñ ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÎπÑÍµê
5. Í≤∞Í≥º Ìï¥ÏÑù: Ïñ¥Îñ§ ÏÉÅÌô©ÏóêÏÑú Ïñ¥Îñ§ ÎåÄÌëúÍ∞íÏù¥ Ï†ÅÏ†àÌïúÏßÄ Ï†ïÎ¶¨

```python
emissions = df['EMISSIONS']

# 1: Í∏∞Î≥∏ ÎåÄÌëúÍ∞í Í≥ÑÏÇ∞ (ÏÇ∞Ïà†ÌèâÍ∑†, Ï§ëÏïôÍ∞í, ÏµúÎπàÍ∞í)
df_mean = emissions.mean()
print(f'ÏÇ∞Ïà†ÌèâÍ∑†: {df_mean:.2f}')
df_median = emissions.median()
print(f'Ï§ëÏïôÍ∞í: {df_median:.2f}')
df_mode = emissions.mode()[0]
print(f'ÏµúÎπàÍ∞í: {df_mode:.2f}')

# 2: Ï†àÏÇ¨ÌèâÍ∑† Í≥ÑÏÇ∞
df_julsa = stats.trim_mean(emissions, 0.1)
print(f'Ï†àÏÇ¨ÌèâÍ∑†: {df_julsa:.2f}')

# 3: Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Ïù∏ Ï∞®ÎüâÏóê ÎåÄÌïú Í∞ÄÏ§ëÌèâÍ∑† Í≥ÑÏÇ∞ (COMPACT 50%, MID-SIZE 30%, SUV 20%)
means = df.groupby('VEHICLE CLASS')['EMISSIONS'].mean()[['COMPACT', 'MID-SIZE', 'SUV']]
weights = np.array([50, 30, 20])
weights_means = np.average(means, weights=weights)
print(f'Í∞ÄÏ§ëÌèâÍ∑†: {weights_means:.2f}')

# TODO 4: Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Î≥Ñ ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÎπÑÍµê
print('='*60)
print("Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Î≥Ñ ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÎπÑÍµê")
print('='*60)
means = df.groupby('VEHICLE CLASS')['EMISSIONS'].mean()[['COMPACT', 'MID-SIZE', 'SUV']]
medians = df.groupby('VEHICLE CLASS')['EMISSIONS'].median()[['COMPACT', 'MID-SIZE', 'SUV']]
print(f'COMPACT: ÌèâÍ∑† {means["COMPACT"]:.2f}, Ï§ëÏïôÍ∞í {medians["COMPACT"]:.2f}')
print(f'MID-SIZE: ÌèâÍ∑† {means["MID-SIZE"]:.2f}, Ï§ëÏïôÍ∞í {medians["MID-SIZE"]:.2f}')
print(f'SUV: ÌèâÍ∑† {means["SUV"]:.2f}, Ï§ëÏïôÍ∞í {medians["SUV"]:.2f}')

# TODO 5: Í≤∞Í≥º Ìï¥ÏÑù (printÎ¨∏ÏúºÎ°ú ÏûëÏÑ±)
```

## üìù Ïã§Ïäµ Î¨∏Ï†ú 2: ÏÇ∞Ìè¨ÎèÑ Ï∏°ÎèÑ

### Î∞∞Í≤Ω
**COMB (L/100 km)** ÏôÄ **ENGINE SIZE**Ïùò ÏÇ∞Ìè¨Î•º Î∂ÑÏÑùÌï©ÎãàÎã§.

### Ïù¥Î°† Î≥µÏäµ

| ÏßÄÌëú | Í≥ÑÏÇ∞ Î∞©Ïãù | Ïù¥ÏÉÅÏπò ÎØºÍ∞êÎèÑ |
|------|----------|:------------:|
| Î∂ÑÏÇ∞/ÌëúÏ§ÄÌé∏Ï∞® | ÌèâÍ∑† Í∏∞Ï§Ä Ìé∏Ï∞® Ï†úÍ≥± | ÎØºÍ∞ê |
| Î≤îÏúÑ | ÏµúÎåÄ - ÏµúÏÜå | Îß§Ïö∞ ÎØºÍ∞ê |
| IQR | Q3 - Q1 | Í∞ïÍ±¥ |

### Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Î∞©Î≤ï
- **IQR Î∞©Î≤ï**: Q1 - 1.5√óIQR ÎØ∏Îßå ÎòêÎäî Q3 + 1.5√óIQR Ï¥àÍ≥º
- **Z-score Î∞©Î≤ï**: |Z| > 2 ÎòêÎäî 3

### ÏàòÌñâÌï† ÏûëÏóÖ
1. Îëê Î≥ÄÏàòÏùò Î∂ÑÏÇ∞, ÌëúÏ§ÄÌé∏Ï∞®, Î≤îÏúÑ, IQR Í≥ÑÏÇ∞
2. Î≥ÄÎèôÍ≥ÑÏàò(CV) Í≥ÑÏÇ∞ÌïòÏó¨ ÏÉÅÎåÄÏ†Å ÏÇ∞Ìè¨ ÎπÑÍµê
3. IQR Î∞©Î≤ïÍ≥º Z-score Î∞©Î≤ïÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Î∞è ÎπÑÍµê
4. 2√ó2 ÏãúÍ∞ÅÌôî: Í∞Å Î≥ÄÏàòÏùò ÌûàÏä§ÌÜ†Í∑∏Îû® + Ï∞®ÎüâÎì±Í∏âÎ≥Ñ Î∞ïÏä§ÌîåÎ°Ø
5. Ï∞®ÎüâÎì±Í∏âÎ≥Ñ(COMPACT, MID-SIZE, SUV) Îëê Î≥ÄÏàòÏùò ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê
6. Í≤∞Í≥º Ìï¥ÏÑù

```python
# 1: Îëê Î≥ÄÏàòÏùò ÏÇ∞Ìè¨ Ï∏°ÎèÑ Í≥ÑÏÇ∞
comb = df['COMB (L/100 km)']
engine_size = df['ENGINE SIZE']

# COMB
print('='*30)
print('COMB:')
print('='*30)
# Î∂ÑÏÇ∞
comb_variance_0 = comb.var(ddof=0)
comb_variance_1 = comb.var(ddof=1)
print(f'Î∂ÑÏÇ∞(Î™®ÏßëÎã®): {comb_variance_0:.4f}')
print(f'Î∂ÑÏÇ∞(ÌëúÎ≥∏ÏßëÎã®): {comb_variance_1:.4f}')
# ÌëúÏ§ÄÌé∏Ï∞®
comb_std_0 = comb.std(ddof=0)
comb_std_1 = comb.std(ddof=1)
print(f'ÌëúÏ§ÄÌé∏Ï∞®(Î™®ÏßëÎã®): {comb_std_0:.4f}')
print(f'ÌëúÏ§ÄÌé∏Ï∞®(ÌëúÎ≥∏ÏßëÎã®): {comb_std_1:.4f}')
# Î≤îÏúÑ
comb_range = comb.max() - comb.min()
print(f'Î≤îÏúÑ: {comb_range:.4f}')

#ENGINE SIZE
engine_size = df['ENGINE SIZE']
print('='*30)
print('ENGINE SIZE:')
print('='*30)
# Î∂ÑÏÇ∞
engine_variance_0 = engine_size.var(ddof=0)
engine_variance_1 = engine_size.var(ddof=1)
print(f'Î∂ÑÏÇ∞(Î™®ÏßëÎã®): {engine_variance_0:.4f}')
print(f'Î∂ÑÏÇ∞(ÌëúÎ≥∏ÏßëÎã®): {engine_variance_1:.4f}')
# ÌëúÏ§ÄÌé∏Ï∞®
engine_std_0 = engine_size.std(ddof=0)
engine_std_1 = engine_size.std(ddof=1)
print(f'ÌëúÏ§ÄÌé∏Ï∞®(Î™®ÏßëÎã®): {engine_std_0:.4f}')
print(f'ÌëúÏ§ÄÌé∏Ï∞®(ÌëúÎ≥∏ÏßëÎã®): {engine_std_1:.4f}')
# Î≤îÏúÑ
engine_range = engine_size.max() - engine_size.min()
print(f'Î≤îÏúÑ: {engine_range:.4f}')

# 2: Î≥ÄÎèôÍ≥ÑÏàò(CV) Í≥ÑÏÇ∞
print('='*30)
print('Î≥ÄÎèôÍ≥ÑÏàò Í≥ÑÏÇ∞')
print('='*30)
comb_cv = (comb.std() / comb.mean()) * 100
print(f'(COMB) CV: {comb_cv:.2f}')
engine_cv = (engine_size.std() / engine_size.mean()) * 100
print(f'(ENGINE SIZE) CV: {engine_cv:.2f}')

# 3: Ïù¥ÏÉÅÏπò ÌÉêÏßÄ (IQR, Z-score)
print('='*30)
print('Ïù¥ÏÉÅÏπò ÌÉêÏßÄ')
print('='*30)
comb_q1 = comb.quantile(0.25)
comb_q3 = comb.quantile(0.75)
comb_iqr = comb_q3 - comb_q1
comb_iqr_lower = comb_q1 - 1.5 * comb_iqr
comb_iqr_upper = comb_q3 + 1.5 * comb_iqr
comb_iqr_outliers = comb[(comb < comb_iqr_lower) | (comb > comb_iqr_upper)]
print('IQR Ïù¥ÏÉÅÏπò:')
print('(COMB) IQR Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(comb_iqr_outliers))
engine_q1 = comb.quantile(0.25)
engine_q3 = comb.quantile(0.75)
engine_iqr = engine_q3 - engine_q1
engine_iqr_lower = engine_q1 - 1.5 * engine_iqr
engine_iqr_upper = engine_q3 + 1.5 * engine_iqr
engine_iqr_outliers = engine_size[(engine_size < engine_iqr_lower) | (engine_size > engine_iqr_upper)]
print('(ENGINE SIZE) IQR Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(engine_iqr_outliers))
comb_z_score = (comb - comb.mean()) / comb.std()
comb_z_outliers_2 = comb[np.abs(comb_z_score) > 2]
comb_z_outliers_3 = comb[np.abs(comb_z_score) > 3]
print('Z-score Ïù¥ÏÉÅÏπò:')
print('(COMB) |Z-score| > 2 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(comb_z_outliers_2))
print('(COMB) |Z-score| > 3 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(comb_z_outliers_3))
engine_z_score = (engine_size - engine_size.mean()) / engine_size.std()
engine_z_outliers_2 = engine_size[np.abs(engine_z_score) > 2]
engine_z_outliers_3 = engine_size[np.abs(engine_z_score) > 3]
print('(ENGINE SIZE) |Z-score| > 2 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(engine_z_outliers_2))
print('(ENGINE SIZE) |Z-score| > 3 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(engine_z_outliers_3))


# 4: 2√ó2 ÏãúÍ∞ÅÌôî: Í∞Å Î≥ÄÏàòÏùò ÌûàÏä§ÌÜ†Í∑∏Îû® + Ï∞®ÎüâÎì±Í∏âÎ≥Ñ Î∞ïÏä§ÌîåÎ°Ø
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# ÌûàÏä§ÌÜ†Í∑∏Îû® - COMB
axes[0, 0].hist(comb, bins=30, alpha=0.7, color='steelblue', edgecolor='black')
axes[0, 0].set_xlabel('COMB (L/100 km)')
axes[0, 0].set_ylabel('ÎπàÎèÑ')
axes[0, 0].set_title(f'COMB (L/100 km) ÌûàÏä§ÌÜ†Í∑∏Îû®\nStd={comb.std():.2f}')
axes[0, 0].grid(True, alpha=0.3)

# ÌûàÏä§ÌÜ†Í∑∏Îû® - ENGINE SIZE
axes[0, 1].hist(engine_size, bins=30, alpha=0.7, color='coral', edgecolor='black')
axes[0, 1].set_xlabel('ENGINE SIZE (L)')
axes[0, 1].set_ylabel('ÎπàÎèÑ')
axes[0, 1].set_title(f'ENGINE SIZE ÌûàÏä§ÌÜ†Í∑∏Îû®\nStd={engine_size.std():.2f}')
axes[0, 1].grid(True, alpha=0.3)

filtered_df = df[df['VEHICLE CLASS'].isin(['COMPACT', 'MID-SIZE', 'SUV'])]

# Ï∞®ÎüâÎì±Í∏âÎ≥Ñ COMB Î∞ïÏä§ÌîåÎ°Ø
sns.boxplot(data=filtered_df, 
            x='VEHICLE CLASS', 
            y='COMB (L/100 km)',
            order=['COMPACT', 'MID-SIZE', 'SUV'],
            palette='Set2',
            ax=axes[1, 0])
axes[1, 0].set_title('Ï∞®ÎüâÎì±Í∏âÎ≥Ñ COMB (L/100 km) Î∂ÑÌè¨')
axes[1, 0].set_xlabel('VEHICLE CLASS')
axes[1, 0].set_ylabel('COMB (L/100 km)')
axes[1, 0].grid(True, alpha=0.3)

# Ï∞®ÎüâÎì±Í∏âÎ≥Ñ ENGINE SIZE Î∞ïÏä§ÌîåÎ°Ø
sns.boxplot(data=filtered_df, 
            x='VEHICLE CLASS', 
            y='ENGINE SIZE',
            order=['COMPACT', 'MID-SIZE', 'SUV'],
            palette='Set3',
            ax=axes[1, 1])
axes[1, 1].set_title('Ï∞®ÎüâÎì±Í∏âÎ≥Ñ ENGINE SIZE Î∂ÑÌè¨')
axes[1, 1].set_xlabel('VEHICLE CLASS')
axes[1, 1].set_ylabel('ENGINE SIZE (L)')
axes[1, 1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 5: Ï∞®ÎüâÎì±Í∏âÎ≥Ñ(COMPACT, MID-SIZE, SUV) ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê - Î∞ïÏä§ÌîåÎ°Ø
print('='*60)
print('Ï∞®ÎüâÎì±Í∏âÎ≥Ñ ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê')
print('='*60)
class_std = filtered_df.groupby('VEHICLE CLASS')[['COMB (L/100 km)', 'ENGINE SIZE']].std()
display(class_std.loc[['COMPACT', 'MID-SIZE', 'SUV']])

# 6: Í≤∞Í≥º Ìï¥ÏÑù
# 1. COMB ÌûàÏä§ÌÜ†Í∑∏Îû®
# - ÏñëÏùò ÏôúÎèÑ Î∂ÑÌè¨
# - ÎåÄÎ∂ÄÎ∂ÑÏùò Ï∞®ÎüâÏù¥ 8~14 L/100km Î≤îÏúÑÏóê ÏßëÏ§ë
# - 20 L/100km Ïù¥ÏÉÅÏùò Ïó∞Î£å ÏÜåÎπÑÍ∞Ä ÌÅ∞ Ï∞®ÎüâÎì§Ïù¥ Ïò§Î•∏Ï™Ω Íº¨Î¶¨Ïóê Ï°¥Ïû¨
# - ÎåÄÎã§Ïàò Ï∞®ÎüâÏùÄ Ïó∞ÎπÑÍ∞Ä ÏñëÌò∏ÌïòÏßÄÎßå, ÏùºÎ∂Ä Ï∞®ÎüâÏù¥ Ïó∞Î£åÎ•º ÎßéÏù¥ ÏÜåÎπÑÌïúÎã§.

# 2. ENGING SIZE ÌûàÏä§ÌÜ†Í∑∏Îû®
# - Îã§Î¥âÎ∂ÑÌè¨? ÏñëÏùò ÏôúÎèÑ Î∂ÑÌè¨?
# - ÎåÄÏ≤¥Î°ú Î∂ÑÌè¨ Î≤îÏúÑÍ∞Ä ÎÑìÏùå
```

## üìù Ïã§Ïäµ Î¨∏Ï†ú 3: Î∂ÑÌè¨Ïùò ÌòïÌÉú (ÏôúÎèÑ, Ï≤®ÎèÑ)

### Î∞∞Í≤Ω
4Í∞ÄÏßÄ ÏàòÏπòÌòï Î≥ÄÏàòÏùò **ÏôúÎèÑÏôÄ Ï≤®ÎèÑ**Î•º Î∂ÑÏÑùÌïòÏó¨ Î∂ÑÌè¨ ÌòïÌÉúÎ•º ÌååÏïÖÌï©ÎãàÎã§.

### Ïù¥Î°† Î≥µÏäµ

**ÏôúÎèÑ**
| ÏôúÎèÑ | ÏùòÎØ∏ | ÌèâÍ∑† vs Ï§ëÏïôÍ∞í |
|------|------|---------------|

```python
# Í≤ÄÏ¶ù ÏΩîÎìú
import numpy as np
from scipy import stats

print("\n" + "="*60)
print("Í≤ÄÏ¶ù Í≤∞Í≥º")
print("="*60)

# ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞
test_data = [4, 7, 2, 9, 1, 5, 3, 8, 6, 5]
test_odd = [3, 1, 4, 1, 5]
test_single = [42]
test_same = [7, 7, 7, 7, 7]
test_negative = [-5, -2, 0, 3, 7]
test_outlier = [1, 2, 3, 4, 100]

all_pass = True
total = 0
passed = 0

def check(name, mine, expected, tol=0.001):
    global all_pass, total, passed
    total += 1
    if mine is None:
        print(f"  ‚ùå {name}: ÎØ∏Íµ¨ÌòÑ")
        all_pass = False
    else:
        try:
            match = abs(float(mine) - float(expected)) < tol
            if match:
                passed += 1
                print(f"  ‚úÖ {name}: {float(mine):.4f}")
            else:
                all_pass = False
                print(f"  ‚ùå {name}: {float(mine):.4f} (Ï†ïÎãµ: {float(expected):.4f})")
        except:
            all_pass = False
            print(f"  ‚ùå {name}: Ïò§Î•ò Î∞úÏÉù")

# 1. ÏÇ∞Ïà†ÌèâÍ∑†
print("\n[my_mean]")
check("Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞", my_mean(test_data), np.mean(test_data))
check("ÌôÄÏàò Í∞ú", my_mean(test_odd), np.mean(test_odd))
check("Îã®Ïùº Í∞í", my_mean(test_single), np.mean(test_single))
check("ÏùåÏàò Ìè¨Ìï®", my_mean(test_negative), np.mean(test_negative))

# 2. Ï†àÏÇ¨ÌèâÍ∑†
print("\n[my_trimmed_mean]")
check("20% Ï†àÏÇ¨", my_trimmed_mean(test_outlier, 0.2), stats.trim_mean(test_outlier, 0.2))
check("10% Ï†àÏÇ¨", my_trimmed_mean(test_data, 0.1), stats.trim_mean(test_data, 0.1))
check("0% Ï†àÏÇ¨", my_trimmed_mean(test_data, 0), stats.trim_mean(test_data, 0))

# 3. Í∞ÄÏ§ëÌèâÍ∑†
print("\n[my_weighted_mean]")
check("ÌïôÏ†ê Í≥ÑÏÇ∞", my_weighted_mean([80, 90, 70], [3, 3, 2]), np.average([80, 90, 70], weights=[3, 3, 2]))
check("ÎèôÏùº Í∞ÄÏ§ëÏπò", my_weighted_mean([100, 200], [1, 1]), np.average([100, 200], weights=[1, 1]))
check("Í∞ÄÏ§ëÏπò Îã§Î¶Ñ", my_weighted_mean([10, 20, 30], [1, 2, 3]), np.average([10, 20, 30], weights=[1, 2, 3]))

# 4. Ï§ëÏïôÍ∞í
print("\n[my_median]")
check("ÏßùÏàò Í∞ú", my_median(test_data), np.median(test_data))
check("ÌôÄÏàò Í∞ú", my_median(test_odd), np.median(test_odd))
check("Îã®Ïùº Í∞í", my_median(test_single), np.median(test_single))
check("ÏùåÏàò Ìè¨Ìï®", my_median(test_negative), np.median(test_negative))

# 5. ÏµúÎπàÍ∞í
print("\n[my_mode]")
check("Í∏∞Î≥∏ (5Í∞Ä 2Î≤à)", my_mode(test_data), stats.mode(test_data, keepdims=False).mode)
check("ÌôÄÏàò (1Ïù¥ 2Î≤à)", my_mode(test_odd), stats.mode(test_odd, keepdims=False).mode)
check("ÎèôÏùº Í∞í", my_mode(test_same), stats.mode(test_same, keepdims=False).mode)

# 6. Î∂ÑÏÇ∞
print("\n[my_variance]")
check("Î™®Î∂ÑÏÇ∞ (ddof=0)", my_variance(test_data, 0), np.var(test_data, ddof=0))
check("ÌëúÎ≥∏Î∂ÑÏÇ∞ (ddof=1)", my_variance(test_data, 1), np.var(test_data, ddof=1))
check("ÎèôÏùº Í∞í ‚Üí 0", my_variance(test_same, 0), np.var(test_same, ddof=0))

# 7. ÌëúÏ§ÄÌé∏Ï∞®
print("\n[my_std]")
check("Î™®ÌëúÏ§ÄÌé∏Ï∞®", my_std(test_data, 0), np.std(test_data, ddof=0))
check("ÌëúÎ≥∏ÌëúÏ§ÄÌé∏Ï∞®", my_std(test_data, 1), np.std(test_data, ddof=1))
check("ÎèôÏùº Í∞í ‚Üí 0", my_std(test_same, 0), np.std(test_same, ddof=0))

# 8. ÌçºÏÑºÌÉÄÏùº (Nearest Rank Î∞©Ïãù)
# Ï£ºÏùò: numpyÎäî ÏÑ†ÌòïÎ≥¥Í∞Ñ ÏÇ¨Ïö©, Nearest RankÎäî ÏßÅÏ†ë Í≥ÑÏÇ∞
print("\n[my_percentile] - Nearest Rank Î∞©Ïãù")
sorted_test = sorted(test_data)  # [1, 2, 3, 4, 5, 5, 6, 7, 8, 9]
check("0% (ÏµúÏÜüÍ∞í)", my_percentile(test_data, 0), min(test_data))
check("50% (Ï§ëÏïôÍ∞í)", my_percentile(test_data, 50), sorted_test[round((len(test_data)-1) * 0.5)])
check("100% (ÏµúÎåìÍ∞í)", my_percentile(test_data, 100), max(test_data))
# 25%, 75%Îäî Nearest Rank Î∞©ÏãùÏúºÎ°ú ÏßÅÏ†ë Í≥ÑÏÇ∞
check("25%", my_percentile(test_data, 25), sorted_test[round((len(test_data)-1) * 0.25)])
check("75%", my_percentile(test_data, 75), sorted_test[round((len(test_data)-1) * 0.75)])

# 9. IQR (Nearest Rank Í∏∞Ï§Ä)
print("\n[my_iqr]")
expected_iqr = sorted_test[round((len(test_data)-1) * 0.75)] - sorted_test[round((len(test_data)-1) * 0.25)]
check("Í∏∞Î≥∏ Îç∞Ïù¥ÌÑ∞", my_iqr(test_data), expected_iqr)
check("ÎèôÏùº Í∞í ‚Üí 0", my_iqr(test_same), 0)

# ÏµúÏ¢Ö Í≤∞Í≥º
print("\n" + "="*60)
print(f"Í≤∞Í≥º: {passed}/{total} ÌÜµÍ≥º ({passed/total*100:.0f}%)")
print("="*60)
if all_pass:
    print("üéâ Î™®Îì† ÌÖåÏä§Ìä∏ ÌÜµÍ≥º!")
else:
    print("‚ö†Ô∏è ÏùºÎ∂Ä ÌÖåÏä§Ìä∏ Ïã§Ìå® - Î°úÏßÅÏùÑ ÌôïÏù∏ÌïòÏÑ∏Ïöî")
```
